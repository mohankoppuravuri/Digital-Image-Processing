Converting RGB image to Gray scale image.

```python
import cv2
import matplotlib.pyplot as plt
actual = cv2.imread(r'C:\Users\Student\Desktop\Test_Images\Lena.tiff')
arr = cv2.cvtColor(actual, cv2.COLOR_BGR2GRAY)
plt.imshow(arr, cmap='gray')
```

```python
def encryption(arrx, ba):
    arr1=arrx.copy()
    rows, colm = arr1.shape
    t=0
    for i in range(rows):
        for j in range(colm):
             if t<len(ba):
                if(ba[t]==False and arr1[i][j]%2 ==1):
                    arr1[i][j]=arr1[i][j]-1
            
                if(ba[t]==True and arr1[i][j]%2 ==0):
                    arr1[i][j]=arr1[i][j]+1
                
                t=t+1  
               
            
    return arr1
```

This is the encrypted image and storing that image as 'encrypted.png' at your current path.


```python
plt.imshow(encryption(arr, ba), cmap='gray')
cv2.imwrite('encrypted.png',encryption(arr, ba))
```
